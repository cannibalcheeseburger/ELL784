```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Path to your AVIRIS .tif file
tif_file = "dataset/aviris_hyperspectral_data/19920612_AVIRIS_IndianPine_EW-line_R.tif"

# Open the GeoTIFF file
with rasterio.open(tif_file) as src:
    # Read the data, assuming multiband (hyperspectral) data
    aviris_data = src.read()  # Shape: (bands, height, width)

# Get some metadata from the file (optional)
print(f"Number of bands: {src.count}")
print(f"Image size (Height x Width): {src.height} x {src.width}")

# Visualize a specific band (e.g., the first spectral band)
band_index = 1  # The band you want to visualize (1-based index for rasterio)
band_data = aviris_data[band_index - 1, :, :]  # Convert to 0-based index for array slicing

# Plot the selected band
plt.imshow(band_data, cmap="viridis")
plt.title(f"AVIRIS Band {band_index}")
plt.colorbar()
plt.show()
```
