```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Load the MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data  # This is the feature matrix (70,000 samples, 784 features)

# Step 2: Standardize the data (mean = 0, variance = 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to find eigenvectors
pca = PCA()  # You can specify n_components to reduce dimensionality if needed
pca.fit(X_scaled)

# Step 4: Get eigenvalues and eigenvectors
eigenvectors = pca.components_  # These are the eigenvectors
eigenvalues = pca.explained_variance_  # These are the corresponding eigenvalues

# Output the first eigenvector and eigenvalue
print("First eigenvector: ", eigenvectors[0])
print("First eigenvalue: ", eigenvalues[0])
```

```python
coefficients = []
for i in range(len(eigenvectors)):
    coefficients.append(np.dot(eigenvalues[i],eigenvectors[i]))
```
