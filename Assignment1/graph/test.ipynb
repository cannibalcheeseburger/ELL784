```python
import matplotlib.pyplot as plt
```

```python
ALPHA_X = [0.001,0.0025,0.005,0.007,0.010]
ALPHA_Y = [0.884,0.9028,0.9188,0.923,0.925]
```

```python
plt.plot(ALPHA_X,ALPHA_Y)
plt.title("Learning Rate vs Validation Accuracy")
plt.xlabel("Learning Rate")
plt.ylabel("Validation Accuracy")
plt.savefig("a_ACC.png")
```

```python
N = [6,7,8,9,10]
L1 = [0.87,0.88,0.894,0.895,0.9]
L2 = [0.88,0.908,0.915,0.925,0.927]
L3 = [0.896,0.913,0.925,0.934,0.939]
L4 = [0.904,0.927,0.928,0.936,0.946]
```

```python
plt.plot(N,L1)
plt.plot(N,L2)
plt.plot(N,L3)
plt.plot(N,L4)
plt.title("Number of Neurons vs Validation Accuracy")
plt.legend(['1', '2',"3","4"], loc='lower right',title="Number of Hidden Layers")
plt.xlabel("Neurons")
plt.ylabel("Validation Accuracy")
plt.savefig("N_Acc.png")
```
