```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from NeuralNetwork.dense import Dense
from NeuralNetwork.activations import Sigmoid,Softmax,reLU
from NeuralNetwork.losses import mse, mse_prime,binary_cross_entropy,binary_cross_entropy_prime
from NeuralNetwork.network import predict
from tqdm import tqdm
```

```python
def train(network, loss, loss_prime, x_train, y_train, epochs = 1000, learning_rate = 0.01, verbose = True):
    error_TS = []
    acc_TS = []
    val_error_TS = []
    val_acc_TS = []
    for e in range(epochs):
        error = 0
        acc =0
        val_error = 0
        val_acc = 0
        for x, y in zip(x_train, y_train):
            # forward
            output = predict(network, x)
            if np.argmax(y)==np.argmax(output):
                acc+=1

            # error

            error += loss(y, output)

            # backward
            grad = loss_prime(y, output)
            for layer in reversed(network):
                grad = layer.backward(grad, learning_rate)

        error /= len(x_train)
        acc /=len(x_train)
        if verbose:
            print(f"{e + 1}/{epochs}, error={error}, accuracy={acc}")
        error_TS.append(error)
        acc_TS.append(acc)
        val_error_TS.append(val_error)
        val_acc_TS.append(val_acc)
    return error_TS,acc_TS
```

```python
df_train = pd.read_csv("./data/train.csv")
df_test = pd.read_csv("./data/test.csv")
```

```python
Y_train,X_train = df_train.iloc[:, :1], df_train.iloc[:, 1:]
Y_test,X_test = df_test.iloc[:, :1], df_test.iloc[:, 1:]
```

```python
X_in = np.reshape(X_train,X_train.shape +(1,))
Y_in = np.squeeze(np.eye(10)[Y_train])
Y_in = np.reshape(Y_in,Y_in.shape +(1,))
```

```python

NN = [Dense(784,40),
      Sigmoid(),
      Dense(40,10),
      Softmax()
      ]

error_TS,acc_TS = train(NN, mse, mse_prime, x_train=X_in,y_train=Y_in, epochs=100, learning_rate=0.01,verbose=False,categorical=True)
```

```python
X_out = np.reshape(X_test,X_test.shape +(1,))
Y_out = np.squeeze(np.eye(10)[Y_test])
Y_out = np.reshape(Y_out,Y_out.shape +(1,))
```

```python
error = 0
acc =0
for x, y in zip(X_out, Y_out):
    output = predict(NN, x)
    error += mse(y, output)
    if np.argmax(y)==np.argmax(output):
        acc+=1
error /= len(X_out)
acc /=len(X_out)
print(f"error={error}, accuracy={acc}")

```
